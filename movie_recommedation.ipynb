```python
def convert_cast(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])

  return L[:5]
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df.head(2)
```

```python
df['crew'][0]
```

```python
def convert_crew(obj):
  M=[]
  for i in ast.literal_eval(obj):

    if i['job']=='Director':
      M.append(i['name'])
  return M
```

```python
df['crew']=df['crew'].apply(convert_crew)
```

```python
def convert_genres(obj):
  G=[]
  for i in ast.literal_eval(obj):
      G.append(i['name'])
  return G
```

```python
df['genres']=df['genres'].apply(convert_genres)
```

```python
def convert_keywords(obj):
  K=[]
  for i in ast.literal_eval(obj):
      K.append(i['name'])
  return K
```

```python
df['keywords']=df['keywords'].apply(convert_keywords)
```

```python

stop_words = set(stopwords.words('english'))
```

```python

lemmatizer = WordNetLemmatizer()
```

```python
def vectorization(list):
    filtered_tokens = [word for word in list if word.lower() not in stop_words]
    lemmas = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    return lemmas
```
